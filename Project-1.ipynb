```python
import pickle
def saveDictionary(tmp_dict):
    try:
        with open('words.txt','wb')as f:
            pickle.dump(tmp_dict,f)
            print('SUCCESS:writing to dictionary words.txt')
    except Exception as e:
        print(f'ERROR:{e}')
def getDictionary():
    try:
        with open('words.txt','rb')as f:
            tmp_dict=pickle.load(f)
            print(tmp_dict)
            print('SUCCESS:reading from dictionary words.txt')
            return tmp_dict
    except Exception as e:
        print(f'ERROR:{e}')
def addNewWord():
    print('*****************')
    print('NEW WORD ADDER:')
    print('*****************')
    print('Enter a Word to add')
    word=input('>>')
    print('Enter a meaning of {word} to add')
    meaning=input('>>')
    dictionary={}
    dictionary[word]=meaning
    saveDictionary(dictionary)
    main_menu()
def findMeaning():
    print('***************')
    print('WORD FINDER:')
    print('***************')
    print('Enter a Word to find')
    word=input('>>')
    dictionary=getDictionary()
    if word in dictionary:
        print(f'Found Entry:[{word}:{dictionary[word]}]')
    else:
        print('ERROR:The word doesnot exist!')
        findMeaning()
    main_menu()
def updateWord():
    print('*************')
    print('WORD UPDATE:')
    print('*************')
    print('Enter an old Word to edit')
    word=input('>>')
    dictionary=getDictionary()
    if word in dictionary:
        print(f'Enter a New Meaning to edit[{word}:{dictionary[word]}]to:')
        new_meaning=input('>>')
        dictionary[word]=new_meaning
        saveDictionary(dictionary)
        print(f'SUCCESS:updated to [{word}:{dictionary[word]}]')
    else:
        print('ERROR:The word doesnot exist!')
        updateWord()
    main_menu()
def main_menu():
    print('**************')
    print('ENGLISH DICTIONARY APP')
    print('**************')
    print('1.Add a new word')
    print('2.Find the meaning')
    print('3.Update a word')
    print('4.Exit')
    print('**************')
    n=int(input('>>'))
    if n==1:
        addNewWord()
    elif n==2:
        findMeaning()
    elif n==3:
        updateWord()
    elif n==4:
        exit()
    else:
        print('ERROR:please Enter a digit according to menu.')
        main_menu()
main_menu()
```

```python
import tkinter
from tkinter import *
import sqlite3 as sq
import tkinter.ttk as ttk
from tkinter import messagebox
root=Tk()
root.title('Libraray Application')
root.geometry('1200x800')
def Database():
    global conn,cursor
    conn=sq.connect('book.db')
    cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS mybook(id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,title TEXT,author TEXT,year INTEGER,isbn INTEGER)')
def Add():
    if TITLE.get()=='' or AUTHOR.get()=='' or YEAR.get()=='':
        txt_result.config(text='Please enter all the fields',fg='blue')
    else:
        Database()
        cursor.execute("INSERT INTO mybook(title,author,year,isbn)VALUES(?,?,?,?)",(TITLE.get(),AUTHOR.get(),YEAR.get(),ISBN.get()))
        conn.commit()
        TITLE.set('')
        AUTHOR.set('')
        YEAR.set('')
        ISBN.set('')
        cursor.close()
        conn.close()
        txt_result.config(text='Record added',fg='green')
def Search():
    if TITLE.get()=='' and AUTHOR.get()=='' and YEAR.get()=='' and ISBN.get()=='':
        txt_result.config(text='Enter details to search',fg='blue')
    else:
        tree.delete(*tree.get_children())
        Database()
        cursor.execute('SELECT * FROM mybook WHERE title=? OR author=? OR year=? OR isbn=?',(TITLE.get(),AUTHOR.get(),str(YEAR.get()),str(ISBN.get())))
        rows=cursor.fetchall()
        for data in rows:
            tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
        TITLE.set('')
        AUTHOR.set('')
        YEAR.set('')
        ISBN.set('')
        cursor.close()
        txt_result.config(text='Search completed and Displayed',fg='green')
def Display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute('SELECT * FROM mybook')
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
    cursor.close()
    txt_result.config(text='Successfully Fetched',fg='blue')
def Exit():
    if messagebox.askokcancel("Quit","Do you want to Exit?"):
        root.destroy()
        exit()
def Delete():
    if TITLE.get()=='' and AUTHOR.get()=='' and YEAR.get()=='' and ISBN.get()=='':
        txt_result.config(text='Please Enter details to delete',fg='blue')
    else:
        Database()
        cursor.execute('DELETE FROM mybook WHERE title=? OR author=? OR year=? OR isbn=?',(TITLE.get(),AUTHOR.get(),YEAR.get(),str(ISBN.get())))
        conn.commit()
        TITLE.set('')
        AUTHOR.set('')
        YEAR.set('')
        ISBN.set('')
        cursor.close()
        conn.close()
        txt_result.config(text='Successfully deleted data',fb='green')
def Issue():
    if TITLE.get()=='' or AUTHOR.get()=='' or YEAR.get()=='':
        txt_result.config(text='Enter all the fields',fg='red')
    else:
        TITLE.set('')
        AUTHOR.set('')
        YEAR.set('')
        ISBN.set('')
        txt_result.config(text='Successfully Issued the book...!',fg='black')
TITLE=StringVar()
AUTHOR=StringVar()
YEAR=StringVar()
ISBN=StringVar()
Top=Frame(root,width=900,height=50,bg='blue',bd=8,relief='raise')
Top.pack(side=TOP)
Left=Frame(root,width=300,height=500,bg='red',bd=8,relief='raise')
Left.pack(side=LEFT)
Right=Frame(root,width=600,height=500,bg='yellow',bd=8,relief='raise')
Right.pack(side=RIGHT)
Forms=Frame(Left,width=300,height=450)
Forms.pack(side=TOP)
Buttons=Frame(Left,width=300,height=100,bd=8,relief='raise')
Buttons.pack(side=BOTTOM)
main_title=Label(Top,width=900,font=('arial',24),text='.....Library Management Application.....')
main_title.pack()
txt_title=Label(Forms,text='BookTitle:',font=('arial',16),bd=15)
txt_title.grid(row=0,stick='e')
txt_author=Label(Forms,text='Author:',font=('arial',16),bd=15)
txt_author.grid(row=1,stick='e')
txt_year=Label(Forms,text='Year:',font=('arial',16),bd=15)
txt_year.grid(row=2,stick='e')
txt_isbn=Label(Forms,text='ISBN:',font=('arial',16),bd=15)
txt_isbn.grid(row=3,stick='e')
txt_result=Label(Buttons)
txt_result.pack()
title=Entry(Forms,textvariable='TITLE',width=30)
title.grid(row=0,column=1)
author=Entry(Forms,textvariable='AUTHOR',width=30)
author.grid(row=1,column=1)
year=Entry(Forms,textvariable='YEAR',width=30)
year.grid(row=2,column=1)
isbn=Entry(Forms,textvariable='ISBN',width=30)
isbn.grid(row=3,column=1)
btn_Display=Button(Buttons,width=10,text='Display',command=Display)
btn_Display.pack(side=LEFT)
btn_Search=Button(Buttons,width=10,text='Search',command=Search)
btn_Search.pack(side=LEFT)
btn_Add=Button(Buttons,width=10,text='Add',command=Add)
btn_Add.pack(side=LEFT)
btn_Issue=Button(Buttons,width=10,text='Issue',command=Issue)
btn_Issue.pack(side=LEFT)
btn_Delete=Button(Buttons,width=10,text='Delete',command=Delete)
btn_Delete.pack(side=LEFT)
btn_Exit=Button(Buttons,width=10,text='Exit',command=Exit)
btn_Exit.pack(side=LEFT)
scrollbary=Scrollbar(Right,orient=VERTICAL)
scrollbarx=Scrollbar(Right,orient=HORIZONTAL)
tree=ttk.Treeview(Right,columns=('BookTitle','Author','Year','ISBN'))
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=RIGHT,fill=X)
tree.heading('BookTitle',text='BookTitle',anchor=W)
tree.heading('Author',text='Author',anchor=W)
tree.heading('Year',text='Year',anchor=W)
tree.heading('ISBN',text='ISBN',anchor=W)
tree.column('#0',stretch=NO,minwidth=0,width=0)
tree.column('#1',stretch=NO,minwidth=0,width=200)
tree.column('#2',stretch=NO,minwidth=0,width=100)
tree.column('#3',stretch=NO,minwidth=0,width=50)
tree.column('#4',stretch=NO,minwidth=0,width=100)
tree.pack()
if __name__ =='__main__':
    root.mainloop()
```
